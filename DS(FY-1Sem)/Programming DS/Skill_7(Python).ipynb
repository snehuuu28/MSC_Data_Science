1. Create python program to design a simple Calculator to perform basic arithmetic
operations like addition, subtraction, multiplication and division.

```python
import tkinter as tk
from tkinter import messagebox

def add():
    try:
        result = float(entry1.get()) + float(entry2.get())
        label_result.config(text="Result: " + str(result))
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter valid numbers.")

def subtract():
    try:
        result = float(entry1.get()) - float(entry2.get())
        label_result.config(text="Result: " + str(result))
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter valid numbers.")

def multiply():
    try:
        result = float(entry1.get()) * float(entry2.get())
        label_result.config(text="Result: " + str(result))
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter valid numbers.")

def divide():
    try:
        denominator = float(entry2.get())
        if denominator == 0:
            messagebox.showerror("Math Error", "Cannot divide by zero.")
            return
        result = float(entry1.get()) / denominator
        label_result.config(text="Result: " + str(result))
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter valid numbers.")

# Create GUI window
window = tk.Tk()
window.title("Simple Calculator")
window.geometry("300x250")
window.configure(padx=10, pady=10)

# Entry fields for numbers
frame_inputs = tk.Frame(window)
frame_inputs.pack(pady=10)

tk.Label(frame_inputs, text="Enter first number:", font=('Arial', 10)).grid(row=0, column=0, pady=5)
entry1 = tk.Entry(frame_inputs, width=20)
entry1.grid(row=0, column=1, pady=5)

tk.Label(frame_inputs, text="Enter second number:", font=('Arial', 10)).grid(row=1, column=0, pady=5)
entry2 = tk.Entry(frame_inputs, width=20)
entry2.grid(row=1, column=1, pady=5)

# Buttons for operations
frame_buttons = tk.Frame(window)
frame_buttons.pack(pady=10)

tk.Button(frame_buttons, text="Add", command=add, width=8).grid(row=0, column=0, padx=5)
tk.Button(frame_buttons, text="Subtract", command=subtract, width=8).grid(row=0, column=1, padx=5)
tk.Button(frame_buttons, text="Multiply", command=multiply, width=8).grid(row=0, column=2, padx=5)
tk.Button(frame_buttons, text="Divide", command=divide, width=8).grid(row=0, column=3, padx=5)

# Label for displaying result
label_result = tk.Label(window, text="Result: ", font=('Arial', 12, 'bold'), pady=10)
label_result.pack()

# Run the application
window.mainloop()
```

2. Create a python program to design login form to accept username and password from
user. Check whether username and password is same or not. If both are same,
display proper message back user.

```python
import tkinter as tk
from tkinter import messagebox

def check_login():
    username = entry_username.get()
    password = entry_password.get()
    if username == password:
        messagebox.showinfo("Login Successful", "Username and password match.")
    else:
        messagebox.showwarning("Login Failed", "Username and password do not match.")

# Create GUI window
window = tk.Tk()
window.title("Login Form")

# Username label and entry field
tk.Label(window, text="Username:").pack()
entry_username = tk.Entry(window)
entry_username.pack()

# Password label and entry field
tk.Label(window, text="Password:").pack()
entry_password = tk.Entry(window, show="*")
entry_password.pack()

# Login button
tk.Button(window, text="Login", command=check_login).pack()

# Run the application
window.mainloop()
```
