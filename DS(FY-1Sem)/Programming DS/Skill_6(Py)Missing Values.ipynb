```python
df.isnull().sum()
```

```python
#Replace Missing value quantity column with  median

df['quantity'].fillna(df['quantity'].median(),inplace=True)
```

```python
#Replace Missing value price column with mean

df['price'].fillna(df['price'].mean(),inplace=True)
```

```python
#Replace Missing value bought column with Standard deviation

df['bought'].fillna(df['bought'].std(),inplace=True)
```

```python
#Replace Missing value forenoon column with max value

df['forenoon'].fillna(df['forenoon'].max(),inplace=True)
```

```python
#Replace Missing value afternoon column with min value

df['afternoon'].fillna(df['afternoon'].min(),inplace=True)
```

2) Create a Python program to load the Data.csv file in DataFrame, find the missingvalues from it. Display which column contains missing values. Replace missing value as follows and display result: (The Data.csv file is uploaded in classroom)
- Replace missing value with some scalar / fix value.
- Replace missing value with value of previous row value
- Replace missing value with value of next row value.
- Replace missing value with value of previous column value
- Replace missing value with value of next row value

```python
df1 = pd.read_csv(r"C:\Users\user\Desktop\MSC(Data Science)\Programing for Data Science\data.csv")
```

```python
df1
```

```python
# Replace missing values with a scalar/fixed value (e.g., 0)
df_scalar = df1.fillna(0)
print("\nDataFrame with Missing Values Replaced by Scalar (0):")
print(df_scalar)
```

```python
# Replace missing values with the value from the previous row (forward fill)
df_previous_row = df1.fillna(method='ffill')
print("\nDataFrame with Missing Values Replaced by Previous Row Value:")
print(df_previous_row)
```

```python
# Replace missing values with the value from the next row (backward fill)
df_next_row = df1.fillna(method='bfill')
print("\nDataFrame with Missing Values Replaced by Next Row Value:")
print(df_next_row)
```

```python
# Replace missing values with the value from the previous column
df_previous_column = df1.fillna(method='ffill', axis=1)

# Display the DataFrame with missing values replaced by the previous column's value
print("\nDataFrame with Missing Values Replaced by Previous Column Value:")
print(df_previous_column)
```

```python
# Replace missing values with the value from the next column (backward fill across columns)
df_next_column = df1.fillna(method='bfill', axis=1)

# Display the DataFrame with missing values replaced by the next column's value
print("\nDataFrame with Missing Values Replaced by Next Column Value:")
print(df_next_column)
```
