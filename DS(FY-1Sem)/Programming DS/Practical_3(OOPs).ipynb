#Q1. Create a class Product with Product ID, Name, Quantity, and Rate as instance variables. 
Accept the values of instance variables from user; calculate the Price of each product and Total Amount.
Display all product details along with Prize per product and Total Amount back to user.

```python
class Product:
    def __init__(self,product_id,product_name,quantity,rate):
        self.product_id=product_id
        self.product_name=product_name
        self.quantity=quantity
        self.rate=rate

    def display(self):
        print("Product_id : ",self.product_id)
        print("Product_name : ",self.product_name)
        print("Quantity : ",self.quantity)
        print("Rate : ",self.rate)
        print("Total Amount : ",self.quantity*self.rate)

product_id=int(input("Enter the product_id : "))
product_name=input("Enter the product_name : ")
quantity=int(input("Enter the quantity : "))
rate=float(input("Enter the rate : "))

p=Product(product_id,product_name,quantity,rate)
p.display()
```

#Q3. Create a class Student with data members roll no & name, add appropriate member methods to accept and display student details.
Inherit 3 classes from Student as Science, Arts & Commerce.
Science class contains marks of two subjects as data members.
Arts class contains marks of two subjects as data members and
Commerce class contains marks of two subjects as data members. 
Create Python program to achieve Hierarchical Inheritance.


```python
class Student:
    def __init__(self,roll,name):
        self.roll=roll
        self.name=name
        
    def display(self):
        print(f"Roll No : {self.roll},Name : {self.name}")

roll=int(input("Enter the roll no : "))
name=input("Enter the name : ")

class Science(Student):
    def __init__(self,roll,name,phy,chem):
        Student.__init__(self,roll,name)
        self.phy=phy
        self.chem=chem
    def display(self):
        print(f"Roll No : {self.roll},Name : {self.name},Physics : {self.phy},Chemistry : {self.chem}")
        
phy=int(input("Enter the physics marks : "))
chem=int(input("Enter the chemistry marks  : "))

class Commerce(Student):
    def __init__(self,roll,name,acc,eco):
        Student.__init__(self,roll,name)
        self.acc=acc
        self.eco=eco
    def display(self):
        print(f"Roll No : {self.roll},Name : {self.name},Accounts : {self.acc},Economics : {self.eco}")
        
acc=int(input("Enter the Accounts marks : "))
eco=int(input("Enter the Economics marks  : "))

class Arts(Student):
    def __init__(self,roll,name,his,geo):
        Student.__init__(self,roll,name)
        self.his=his
        self.geo=geo
    def display(self):
        print(f"Roll No : {self.roll},Name : {self.name},History : {self.his},Geography : {self.geo}")
        
his=int(input("Enter the History marks : "))
geo=int(input("Enter the Geography marks : "))

# Create instances for each class
science_student = Science(roll, name, phy, chem)
commerce_student = Commerce(roll, name, acc, eco)
arts_student = Arts(roll, name, his, geo)

# Display information for each student
print("\nScience Student Details:")
science_student.display()

print("\nCommerce Student Details:")
commerce_student.display()

print("\nArts Student Details:")
arts_student.display()
```

#Q4. Create an abstract base class called Shape with two instance variables Dimention-1 and Dimention-2. 
Dimensions of shape used to compute the area. 
Derive two specific classes called Triangle and 
Rectangle from the base Shape. 
Call the base class constructor through derived class constructor for initializing data members.
Add one member method area () to compute and display area of shapes. 
Create python program to implement run time polymorphism by using method overriding.

```python
from abc import ABC,abstractmethod
class Shape(ABC):
    def __init__(self,d1,d2):
        self.d1=d1
        self.d2=d2
    def area(self):
        pass

class Triangle(Shape):
    def __init__(self,base,height):
        Shape.__init__(self,base,height)
    def area(self):
        return 0.5*self.d1*self.d2

class Rectangle(Shape):
    def __init__(self,lenght,breadth):
        Shape.__init__(self,lenght,breadth)
    def area(self):
        return self.d1*self.d2

T=Triangle(5,6)
R=Rectangle(5,3)

print("Triangle : ",T.area())
print("Rectangle : ",R.area())
        
```
