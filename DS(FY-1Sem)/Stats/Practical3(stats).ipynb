```python
# 7) Out of set of independent variables, identify the variables which are linearly related to each other.

correlation_matrix = df.corr()
threshold = 0.5
# Identify pairs of highly correlated variables
highly_correlated_pairs = [(correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i, j])
                           for i in range(len(correlation_matrix.columns))
                           for j in range(i) 
                           if abs(correlation_matrix.iloc[i, j]) > threshold]

# Print the highly correlated pairs
print("Highly correlated variable pairs (|correlation| > 0.5):")
for pair in highly_correlated_pairs:
    print(f"{pair[0]} and {pair[1]}: {pair[2]:.2f}")

# Visualize the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Independent Variables")
plt.tight_layout()
plt.show()
```

```python
# 8) Find the correlation between bodyfat and linear combination of weight and height. Comment on your  finding. 
```

```python
linear_combination = df['WEIGHT'] + df['HEIGHT']
# Calculate the correlation between BODYFAT and the linear combination
correlation = df['BODYFAT'].corr(linear_combination)
# Print the correlation result
print("Correlation between BODYFAT and linear combination of WEIGHT and HEIGHT:", correlation)
# Comment on findings
if correlation > 0.5:
    correlation_comment = "There is a strong positive correlation between BODYFAT and the linear combination of WEIGHT and HEIGHT."
elif correlation < -0.5:
    correlation_comment = "There is a strong negative correlation between BODYFAT and the linear combination of WEIGHT and HEIGHT."
else:
    correlation_comment = "The correlation between BODYFAT and the linear combination of WEIGHT and HEIGHT is weak or moderate."
print("\nComment:", correlation_comment)
```
