# Step 1: Load the Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\user\Desktop\MSC(Data Science)\Statistic\statistic(Python)\NationalNames.csv")
```

```python
df
```

# Step 2: Visualize the Number of Male and Female Babies Born in Each Year

```python
#Group the data by year and gender:
gender_counts = df.groupby(['Year', 'Gender'])['Count'].sum().unstack()
```

```python
#Plot the results
gender_counts.plot(kind='bar', figsize=(12, 6), title='Number of Male and Female Babies Born per Year')
```

# Step 3: Analyze Popular Baby Names

```python
#Group by name and sum the counts:
popular_names = df.groupby('Name')['Count'].sum().reset_index()
```

```python
#Sort to find the top 100 baby names:
top_100_names = popular_names.sort_values(by='Count', ascending=False).head(100)
```

```python
#Display the top 100 popular baby names:
print(top_100_names)
```

# Step 4: Visualize the Top 10 Popular Baby Names

```python
top_10_names = top_100_names.head(10)
top_10_names.plot(kind='bar', x='Name', y='Count', title='Top 10 Popular Baby Names', figsize=(10, 5))
```
