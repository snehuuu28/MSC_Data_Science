```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

#plot settings 
plt.figure(figsize=(18,5))

#Norm Distribution
plt.subplot(1,3,1)
mu,sigma=0,1       #mean and SD
x=np.linspace(-5,5,1000)
plt.plot(x,stats.norm.pdf(x,mu,sigma), label=f'$\mu=(mu), \sigma=(sigma)$')
plt.title("Normal Distribution")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.legend()
```

```python
#Binomial Distribution
plt.subplot(1,3,2)
n,p=10,0.5      
x=np.arange(0,n+1)
plt.stem(x,stats.binom.pmf(x,n,p), basefmt=" ")
plt.title("Binomial Distribution")
plt.xlabel("X")
plt.ylabel("Probability Mass Function")
plt.show()
```

```python
#Poisson Distribution

plt.subplot(1,3,3)
lambda_=3    #rate parameter
x=np.arange(0,15)
plt.stem(x,stats.poisson.pmf(x,lambda_))
plt.title("Poisson Distribution")
plt.xlabel("X")
plt.ylabel("Probability Mass Function")
plt.tight_layout()
plt.show()
```
