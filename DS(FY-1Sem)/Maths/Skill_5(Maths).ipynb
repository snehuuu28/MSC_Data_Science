## 1) Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/ (1+np.exp(-x))

# Example usage
x_values = np.linspace(-7, 7, 100)  # Generate 100 values from -7 to 7
sigmoid_values = sigmoid(x_values)

# Plotting the sigmoid function
plt.plot(x_values, sigmoid_values, label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## 2) SoftMax Function

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)

# Generate values for the x-axis
x_values = np.linspace(-5, 5, 100)

# Calculate softmax values for each x value
softmax_values = np.array([softmax(x) for x in x_values])

# Plotting the softmax function
plt.plot(x_values, softmax_values, label='Softmax Function')
plt.title('Softmax Function')
plt.xlabel('x')
plt.ylabel('softmax(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## 3) Loss Function

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a range for prediction errors
errors = np.linspace(-2, 2, 400)  # Range of prediction errors
delta = 1.0  # Delta for Huber loss

# Mean Squared Error (MSE) Loss
mse_loss = errors ** 2

# Mean Absolute Error (MAE) Loss
mae_loss = np.abs(errors)

# Huber Loss
huber_loss = np.where(np.abs(errors) <= delta, 
                      0.5 * errors ** 2, 
                      delta * (np.abs(errors) - 0.5 * delta))

# Plotting
plt.figure(figsize=(5, 3))
plt.plot(errors, mse_loss, label="MSE Loss", color="blue")
plt.plot(errors, mae_loss, label="MAE Loss", color="green")
plt.plot(errors, huber_loss, label="Huber Loss", color="red")
plt.xlabel("Prediction Error")
plt.ylabel("Loss")
plt.title("Comparison of Loss Functions")
plt.legend()
plt.grid()
plt.show()
```

## 4) ReLU Function

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

# Example usage
x_values = np.linspace(-5, 5, 100)
relu_values = relu(x_values)

# Plotting the ReLU function
plt.plot(x_values, relu_values, label='ReLU Function')
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## 5) TanH Function

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of x values
x = np.linspace(-5, 5, 400)

# Calculate tanh for each x value
tanh_values = np.tanh(x)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, tanh_values, label="tanh(x)", color="purple")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.title("Hyperbolic Tangent Function")
plt.legend()
plt.grid()
plt.show()
```
